# Exploratory Data Analysis

```python
plt.figure(figsize=(6,5))
sns.kdeplot(df['Sales'],color='red',label='SALES',shade=True)
sns.kdeplot(df['Profit'],color='Blue',label='PROFIT', shade=True)
plt.xlim([-100,1000])
plt.legend()
plt.show()
```

Here, we can see that Profit is more tham sales but in some areas profit can still be increased.

# Statewise deal Analysis

```python
df1 = df['State'].value_counts()
df1.head(10)
```

```python
df1.plot(kind='bar',figsize=(15,5))
plt.ylabel('Frequency / Number of deals')
plt.xlabel('States')

plt.title('State Wise Dealings', fontsize = 20)
plt.show()
```

 states with highest deals are California, New York and Texas.
 state with least deals is Wyoming

```python
df['State'].value_counts().mean()
```

Average number of deal per State is 204.

# City Wise analysis of the dealing

```python
df2 = df['City'].value_counts()
df2=df2.head(50)
df2.plot(kind='bar',figsize=(15,5))
plt.ylabel('Frequency / Number of deals')
plt.xlabel('City')

plt.title('City Wise Dealings', fontsize = 20)
plt.show()
```

New York city gets highest deals whereas Providence gets the least deal.

```python
df['City'].value_counts().mean()
```

Average number of deal per city is 19

### Statewise analysis of Profit Discount and sales

```python
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state
```

```python
#STATEWISE ANALYSIS OF PROFIT

df_state1=df_state.sort_values('Profit')

df_state1[['Profit']].plot(kind = 'bar', color='green' ,figsize = (12,5))
plt.title('State wise Profit Analysis', fontsize = 30)
plt.ylabel('Profit per Sate')
plt.xlabel('States')
plt.show()
```

VERMONT has the highest profit while OHIO has the lowest profit

```python
# STATEWISE ANALYSIS OF SALES


df_state[["Sales"]].plot(kind = 'bar',color="orange" ,figsize = (15,4))
plt.title('State wise sales Analysis', fontsize = 20)
plt.ylabel('Sales per Sate')
plt.xlabel('States')
plt.show()
```

WYOMING has the highest amount of sales whereas South Dakota has the lowest amount of Sales

```python
# STATEWISE ANALYSIS OF DISCOUNT

df_state1['Discount'].plot(kind='bar',figsize=(15,5))
plt.title('State wise analysis of Discount', fontsize=20)
```

### CATEGORY WISE SALES DISCOUNT AND PROFIT 

```python
df_category = df.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
df_category
```

```python
#sales
#Discount
#Profit

df_category.plot.pie(subplots=True, 
                     figsize=(18, 20), 
                     autopct='%1.1f%%', 
                     labels = df_category.index)
```

-Maximum sales and profit comes from Technology
-Minimum Profit is obtained from Furniture

### Sub-Category wise Sales, Profit and Discount 

```python
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category
```

#### 1) BASED ON THE SALES 

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Copier and Machine have High sales

#### 2) BASED ON THE DISCOUNT 

```python
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 30)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Binders , Machines and tables have highest Discount.

#### 3) BASED ON THE PROFIT

```python
df_sub_category.sort_values('Profit')[['Sales','Profit']].plot(kind='bar',figsize= (10,5),label=['Avg Sales Price($)','Profit($)'])
```
